```python
import pysolr

# 连接到Solr
solr = pysolr.Solr('http://127.0.0.1:8983/solr/bioner-diseases', timeout=10)

# 根据你的schema，只有这两个字段可以搜索：
# - term (indexed=true)
# - synonyms (indexed=true)
# 其他字段如 mesh_id, cui, semantic_type 等不能用于搜索

results = solr.search('term:test', rows=5)

print(f"找到 {len(results)} 条结果:")
for doc in results:
    print(f"术语: {doc.get('term', 'N/A')}")
    print(f"同义词: {doc.get('synonyms', [])}")
    print(f"Mesh ID: {doc.get('mesh_id', 'N/A')}")

# 示例2: 同时在term和synonyms字段中搜索
results = solr.search('term:diabetes OR synonyms:diabetes', rows=3)

# 示例3: 查询所有文档
all_docs = solr.search('*:*', rows=10)
```
